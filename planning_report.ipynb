```r
library(tidyverse)

players <- read_csv("players.csv")
sessions <- read_csv("sessions.csv")

dim(players); head(players)
dim(sessions); head(sessions)
```

```r
# variable names
names(players)
names(sessions)
```

```r
# variable types and quick preview
glimpse(players)
glimpse(sessions)
```

```r
distinct(players, experience)
distinct(players, subscribe)
distinct(players, gender)

players |>
  summarise(
    max_played_hours = max(played_hours, na.rm = TRUE),
    min_played_hours = min(played_hours, na.rm = TRUE),
    max_age = max(Age, na.rm = TRUE),
    min_age = min(Age, na.rm = TRUE)
  )
```

```r
players %>%
  summarise(across(where(is.numeric), ~ round(mean(.x, na.rm = TRUE), 2)))

sessions %>%
  summarise(across(where(is.numeric), ~ round(mean(.x, na.rm = TRUE), 2)))
```

# This project uses two datasets: players.csv and sessions.csv.


## 1. Number of observations & variables
	•	players.csv: 196 rows, 7 columns
	•	sessions.csv: 1535 rows, 5 columns

## 2. Variable names

players.csv: experience, subscribe, hashedEmail, played_hours, name, gender, Age
sessions.csv: hashedEmail, start_time, end_time, original_start_time, original_end_time

## 3. Variable types & meanings
	•	experience (chr): player skill level
	•	subscribe (lgl): subscribed or not
	•	hashedEmail (chr): anonymized ID
	•	played_hours (dbl): total hours played
	•	name (chr): chosen player name
	•	gender (chr): gender identity
	•	Age (dbl): player age
	•	start_time / end_time (chr): gameplay session time
	•	original_start_time / original_end_time (dbl): numeric timestamps

## 4. Unique categories
	•	experience: Pro, Veteran, Amateur, Regular, Beginner
	•	subscribe: TRUE, FALSE
	•	gender: Male, Female, Non-binary, Prefer not to say, Agender, Two-Spirited, Other
