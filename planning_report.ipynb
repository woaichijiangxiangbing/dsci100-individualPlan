## Broad Question

## Specific Question  
**Can a player's characteristics (experience, played hours, gender, and age) predict whether they subscribe to the newsletter?**

The `players.csv` dataset contains all variables needed to answer this question, including:
- `subscribe` (response)
- `experience`
- `played_hours`
- `gender`
- `Age`

These variables allow us to examine relationships between player behaviours/demographics and subscription outcomes.

## Summary statistics (mean values)

Below I compute the mean value for each quantitative variable in the players.csv dataset.

```r
library(dplyr)

players %>%
  summarise(
    mean_played_hours = mean(played_hours, na.rm = TRUE),
    mean_age = mean(Age, na.rm = TRUE)
  )
```

```r
ggplot(players, aes(x = experience, fill = subscribe)) +
  geom_bar(position = "fill") +
  labs(
    title = "Subscription rate by Experience Level",
    x = "Experience Level",
    y = "Proportion"
  )
```

```r
ggplot(players, aes(x = experience, y = played_hours)) +
  geom_boxplot() +
  labs(
    title = "Played Hours by Experience Level",
    x = "Experience Level",
    y = "Played Hours"
  )
```

```r
ggplot(players, aes(x = Age, fill = subscribe)) +
  geom_histogram(position = "identity", alpha = 0.5, bins = 20) +
  labs(title = "Age Distribution by Subscription Status")
```

### Final Insights

- Experience level seems related to newsletter subscription: more experienced players show slightly higher subscription proportions.  
- Played hours vary widely within each experience level, suggesting some players play very intensively while others play very little.  
- Age is concentrated in younger players, and subscription patterns are fairly similar across most age groups. These patterns help identify useful variables and potential issues (e.g., outliers or skewed distributions) for later modelling.
